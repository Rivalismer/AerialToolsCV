```python
import os
import shutil
import skimage
from skimage import io
```

```python
# Set paths - Assumes folder structure as given by the dataset
old_path = "../PV01/"

img_train_path_new = "../Data/pv01/train/images"
img_test_path_new = "../Data/pv01/test/images"
img_validation_path = "../Data/pv01/validation/images"

label_train_path_new = "../Data/pv01/train/labels"
label_test_path_new = "../Data/pv01/test/labels"
label_validation_path = "../Data/pv01/validation/labels"
```

```python
# Find images
img = []
labels = []
subfld = [f for f in os.listdir(old_path)]

for fold in subfld:
    temp_img = [f for f in os.listdir(old_path + fold) if not '_label' in f.lower()]
    temp_labels = [f for f in os.listdir(old_path + fold) if '_label' in f.lower()]
    img.extend(temp_img)
    labels.extend(temp_labels)

img_validation = img[::4]
img_test = img[::7]

label_validation = labels[::4]
labels_test = labels[::7]

for f in img_validation:
    if f in img:
        img.remove(f)
    if f in img_test:
        img_test.remove(f)

for f in label_validation:
    if f in labels:
        labels.remove(f)
    if f in labels_test:
        labels_test.remove(f)
```

```python
# Move to folder structure

# Images
for fold in subfld:
    for imgs in img:
        if imgs in os.listdir(old_path + fold):
            old = old_path + fold + '/' + imgs
            img_path = img_train_path_new + imgs
            shutil.move(old, img_path)

    for imgs in img_test:
        if imgs in os.listdir(old_path + fold):
            old = old_path + fold + '/' + imgs
            img_path = img_test_path_new + imgs
            shutil.move(old, img_path)

    for imgs in img_validation:
        if imgs in os.listdir(old_path + fold):
            if img in img_validation:
                old = old_path + fold + '/' + imgs
            else:
                old = old_path + fold + '/' + imgs
            img_path = img_validation_path + imgs
            shutil.move(old, img_path)

    # Labels
    for imgs in labels:
        if imgs in os.listdir(old_path + fold):
            old = old_path + fold + '/' + imgs
            img_path = label_train_path_new + imgs
            shutil.move(old, img_path)

    for imgs in labels_test:
        if imgs in os.listdir(old_path + fold):
            old = old_path + fold + '/' + imgs
            img_path = label_test_path_new + imgs
            shutil.move(old, img_path)

    for imgs in label_validation:
        if imgs in os.listdir(old_path + fold):
            if img in labels:
                old = old_path + fold + '/' + imgs
            else:
                old = old_path + fold + '/' + imgs
            img_path = label_validation_path + imgs
            shutil.move(old, img_path)
```

```python
# In case the test images and labels went unmoved
for imgs in img_test:
    old = img_train_path_new + imgs
    new = img_test_path_new + imgs
    shutil.move(old, new)

for label in labels_test:
    old = label_train_path_new + label
    new = label_test_path_new + label
    shutil.move(old,new)
```

```python
# Ensure that they are still sorted correctly
train_new = [f for f in os.listdir(img_train_path_new)]
labels_new = [f for f in os.listdir(label_train_path_new)]

try:
    j = 0
    for i in range(len(train_new)):
        # if i >= 25:
        #     tp = train_new[i].removesuffix(".JPG")
        #     lp = labels_new[i+1].removesuffix(".png")
        #     #print(f"tp: {tp}\n lp:{lp}")
        #     if tp == lp:
        #         j += 1
        #     else:
        #         print(f"tp: {tp} \n lp:{lp}")
        #         print(f"Stopped at: {j}, incorrect sorting")
        tp = train_new[i].removesuffix(".bmp")
        lp = labels_new[i].removesuffix(".bmp")
        #print(f"tp: {tp}\nlp: {lp}")
        if (tp + '_label') == lp:
            j += 1
        else:
            print(f"Stopped at: {j}, incorrect sorting")
except:
    pass
```
