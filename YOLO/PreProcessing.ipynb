```python
import os
import shutil
import skimage
from skimage import io
```

```python
# Set paths - Assumes folder structure as given by the dataset
img_train_path_old = "../imgs/train_val_set/"
img_test_path_old = "../imgs/test_set/"

label_train_path_old = "../imgs/train_val_set/"
label_test_path_old = "../imgs/test_set/"

img_train_path_new = "../Data/img/train/"
img_test_path_new = "../Data/img/test/"
img_validation_path = "../Data/img/validation/"

label_train_path_new = "../Data/labels/train/"
label_test_path_new = "../Data/labels/test/"
label_validation_path = "../Data/labels/validation/"
```

```python
# Find images
img_train = [f for f in os.listdir(img_train_path_old) if '.jpg' in f.lower()]
img_test = [f for f in os.listdir(img_test_path_old) if '.jpg' in f.lower()]
img_validation_train = img_train[::4]
img_validation_test = img_test[::4]
img_validation = img_validation_train + img_validation_test

label_train = [f for f in os.listdir(label_train_path_old) if '.png' in f.lower()]
label_test = [f for f in os.listdir(label_test_path_old) if '.png' in f.lower()]
label_validation_train = label_train[::4]
label_validation_test = label_test[::4]
label_validation = label_train[::4] + label_test[::4]

for f in img_validation:
    if f in img_train:
        img_train.remove(f)
    elif f in img_test:
        img_test.remove(f)

for f in label_validation:
    if f in label_train:
        label_train.remove(f)
    elif f in label_test:
        label_test.remove(f)
```

```python
# Move to folder structure

# Images
for img in img_train:
    old_path = img_train_path_old + img
    img_path = img_train_path_new + img
    shutil.move(old_path, img_path)

for img in img_test:
    old_path = img_test_path_old + img
    img_path = img_test_path_new + img
    shutil.move(old_path, img_path)
    
for img in img_validation:
    if img in img_validation_train:
        old_path = img_train_path_old + img
    else:
        old_path = img_test_path_old + img
    img_path = img_validation_path + img
    shutil.move(old_path, img_path)
    
# Labels
for img in label_train:
    old_path = label_train_path_old + img
    img_path = label_train_path_new + img
    shutil.move(old_path, img_path)

for img in label_test:
    old_path = label_test_path_old + img
    img_path = label_test_path_new + img
    shutil.move(old_path, img_path)
    
for img in label_validation:
    if img in label_validation_train:
        old_path = label_train_path_old + img
    else:
        old_path = label_test_path_old + img
    img_path = label_validation_path + img
    shutil.move(old_path, img_path)
```
