```python
import os
import shutil
import numpy as np
from PIL import Image, ImageOps
from skimage.measure import label
```

```python
# Paths and labels
label_train_path_new = "../Data/labels/train/"
label_test_path_new = "../Data/labels/test/"
label_validation_path = "../Data/labels/validation/"

train_mask_paths = [os.path.join(label_train_path_new, f) for f in os.listdir(label_train_path_new) if ".png" in f.lower()]
test_mask_paths = [os.path.join(label_test_path_new, f) for f in os.listdir(label_test_path_new) if ".png" in f.lower()]
validation_mask_paths = [os.path.join(label_validation_path, f) for f in os.listdir(label_validation_path) if ".png" in f.lower()]

# train_names = [f.removesuffix(".png") for f in os.listdir(label_train_path_new)]
# test_names = [f.removesuffix(".png") for f in os.listdir(label_test_path_new)]
# validation_names = [f.removesuffix(".png") for f in os.listdir(label_validation_path)]
```

```python
# Might need to resize image to 640 x 640 - thus normalization of the bounding boxes would be needed.
# Currently all images are 2250 x 4000
for i in range(len(train_mask_paths)):
    # Get name of txt and path
    path = train_mask_paths[i]
    name = path.removesuffix(".png") + ".txt"
    #print(name)
    # Open and convert to grayscale
    mask = Image.open(path)
    mask = ImageOps.grayscale(mask)

    # Convert PIL to np-array
    mask = np.array(mask)
    label_im = label(mask)
    vals = np.unique(label_im)
    vals = vals[1:]
    masks = (label_im[:, None, None] == vals[:, None, None])
    masks = np.transpose(masks, (2,1,0,3))

    num_objs = len(vals)
    center_x = []
    center_y = []
    w = []
    h = []
    area = []
    boxes = []
    for i in range(num_objs):
        pos = np.where(label_im == vals[i])
        xmin = np.min(pos[1])/2250
        xmax = np.max(pos[1])/2250
        ymin = np.min(pos[0])/4000
        ymax = np.max(pos[0])/4000
        boxes.append([xmin, ymin, xmax, ymax])
        x = xmin + (xmax - xmin)//2
        y = ymin + (ymax - ymin)//2
        center_x.append(x)
        center_y.append(y)
        h.append(ymax - ymin)
        w.append(xmax - xmin)
        area.append((ymax - ymin) * (xmax - xmin))

    with open(name, "a") as f:
        for i in range(num_objs):
            content = '0 ' + str(center_x[i]) + ' ' + str(center_y[i]) + ' ' + str(w[i]) + ' ' + str(h[i])
            #print(content)
            f.write(content)
```

```python
# Move .txts
test = [f for f in os.listdir(label_test_path_new) if ".txt" in f.lower()]
train = [f for f in os.listdir(label_train_path_new) if ".txt" in f.lower()]
validation = [f for f in os.listdir(label_validation_path) if ".txt" in f.lower()]

yolov7_train_path = "../Data/yolov7/labels/train/"
yolov7_test_path = "../Data/yolov7/labels/test/"
yolov7_validation_path = "../Data/yolov7/labels/validation/"

```

```python
for label in train:
    old = label_train_path_new + label
    new = yolov7_train_path + label
    shutil.move(old, new)

for label in test:
    old = label_test_path_new + label
    new = yolov7_test_path + label
    shutil.move(old, new)

for label in validation:
    old = label_validation_path + label
    new = yolov7_validation_path + label
    shutil.move(old,new)
```
