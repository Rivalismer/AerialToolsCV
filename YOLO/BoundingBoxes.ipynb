```python
import os
import shutil
import numpy as np
from PIL import Image, ImageOps
from skimage.measure import label
from skimage import io
```

```python
# Paths and labels
label_train_path_new = "../Data/labels/train/"
label_test_path_new = "../Data/labels/test/"
label_validation_path = "../Data/labels/validation/"

train_mask_paths = [os.path.join(label_train_path_new, f) for f in os.listdir(label_train_path_new) if ".png" in f.lower()]
test_mask_paths = [os.path.join(label_test_path_new, f) for f in os.listdir(label_test_path_new) if ".png" in f.lower()]
validation_mask_paths = [os.path.join(label_validation_path, f) for f in os.listdir(label_validation_path) if ".png" in f.lower()]

# train_names = [f.removesuffix(".png") for f in os.listdir(label_train_path_new)]
# test_names = [f.removesuffix(".png") for f in os.listdir(label_test_path_new)]
# validation_names = [f.removesuffix(".png") for 
```

```python
# Might need to resize image to 640 x 640 - thus normalization of the bounding boxes would be needed.
# Currently all images are 2250 x 4000
for i in range(len(train_mask_paths)):
    # Get name of txt and path
    path = train_mask_paths[i]
    name = path.removesuffix(".png") + ".txt"
    
    # Open and convert to grayscale
    mask = Image.open(path)
    mask = ImageOps.grayscale(mask)
    x_norm, y_norm = mask.size
    # Convert PIL to np-array
    mask = np.array(mask)
    label_im = label(mask)
    vals = np.unique(label_im)
    vals = vals[1:]
    masks = (label_im[:, None, None] == vals[:, None, None])
    masks = np.transpose(masks, (2,1,0,3))

    num_objs = len(vals)
    center_x = []
    center_y = []
    w = []
    h = []
    boxes = []
    for i in range(num_objs):
        pos = np.where(label_im == vals[i])
        xmin = np.min(pos[1])/x_norm
        xmax = np.max(pos[1])/x_norm
        ymin = np.min(pos[0])/y_norm
        ymax = np.max(pos[0])/y_norm
        boxes.append([xmin, ymin, xmax, ymax])
        x = xmin + (xmax - xmin)/2
        y = ymin + (ymax - ymin)/2
        if x > 1 or y > 1:
            print("Error in normalisation")
            print(f"X: {x}, Y: {y}")
        center_x.append(x)
        center_y.append(y)
        h.append(ymax - ymin)
        w.append(xmax - xmin)

    with open(name, "a") as f:
        for i in range(num_objs):
            if i != num_objs - 1:
                content = '0 ' + str(center_x[i]) + ' ' + str(center_y[i]) + ' ' + str(w[i]) + ' ' + str(h[i]) + '\n'
                #print(content)
                f.write(content)
            else:
                content = '0 ' + str(center_x[i]) + ' ' + str(center_y[i]) + ' ' + str(w[i]) + ' ' + str(h[i])
                #print(content)
                f.write(content)
```
